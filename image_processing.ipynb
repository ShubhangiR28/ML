```python
import numpy as np
from skimage import io,draw
import matplotlib.pyplot as plt 

image = io.imread('teddy.jpg')
plt.imshow(image)

```

```python
mask = np.ones_like(image)*255
w,h,_ = image.shape 

circle_center = (w//2,h//2)
circle_radius = min(w,h) // 2


rr, cc = draw.disk(center=circle_center, radius=circle_radius, shape=(w,h))

mask[rr,cc,:] = image[rr,cc,:]
plt.imshow(mask.astype(np.uint8))
plt.axis('off')
plt.show()
```
